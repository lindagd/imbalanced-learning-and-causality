# Criando datasets sintéticos para classificação binária

```python
from sklearn.datasets import make_classification
```

## Criando e exibindo um dataset balanceado

```python
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=1)
```

```python
plt.scatter(X[:,0], X[:, 1], c=y, edgecolors='k')
```

## Criando e exibindo um dataset desbalanceado

```python
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_clusters_per_class=1,
                           weights=[0.9], random_state=1)
```

```python
from collections import Counter
import numpy as np
import csv

c = Counter(y)
print(c)

plt.scatter(X[:,0], X[:, 1], c=y, edgecolors='k')
#print(f'Tamanho = {len(X)} Proporção = {int(len(X)*0.99)}/{int(len(X)*0.01)}')

def export_csv(name, X, y):
  with open(name+'.csv', 'w', newline='') as file:
      writer = csv.writer(file)
      field = ["x coordenate", "y coordenate", "class"]
      writer.writerow(field)
      i = 0
      for point in X:
        s = f'{point}'
        row = s.strip('[]').split()
        row.append(y[i])
        writer.writerow(row)
        i += 1

export_csv('teste', X, y)

#print(X)
print(y)
```
