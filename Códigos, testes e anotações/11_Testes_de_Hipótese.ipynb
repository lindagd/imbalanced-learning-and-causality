```python
# importar aurocs já calculadas
import scores

scores = scores.auroc_scores

# importar tipos de testes de hipóteses da biblioteca scipy
from scipy.stats import mannwhitneyu
from scipy.stats import kruskal

#import numpy as np
from numpy import array
import csv
```

```python
aurocs = list(scores.values())

h_stat, p_val = kruskal(*aurocs)
results_kw = [h_stat, p_val]

print(results_kw)
```

```python
with open('results_mannwhitney.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    field = ["pair", "statistic", "p-value"]
    writer.writerow(field)

results_mw = {}

with open('results_MW_low_p_val.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    field = ["pair", "statistic", "p-value"]
    writer.writerow(field)
mw_low_p_val = {}

for technique, scores1 in scores.items():
  results_mw[technique] = {}
  for t, s in scores.items():
    if t not in results_mw:
      stat, p_val = mannwhitneyu(scores1, s)
      title = technique + ' VS ' + t
      results_mw[technique][t] = [stat, p_val]
      row = [title, stat, p_val]

      with open('results_mannwhitney.csv', 'a') as f_object:
        writer_obj = csv.writer(f_object)
        writer_obj.writerow(row)
        f_object.close()

      if p_val < 0.05:
        results_mw[title] = [stat, p_val]
        with open('results_MW_low_p_val.csv', 'a') as f_object:
          writer_obj = csv.writer(f_object)
          writer_obj.writerow(row)
          f_object.close()
```
